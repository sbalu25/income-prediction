```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
```

```python
data = pd.read_table('adult.data',sep = ',',header=None)
```

```python
columns = ['age','workclass','fnlwgt','education','education-num','marital-status',\
          'occupation','relationship','race','sex','capital-gain',\
          'capital-loss','hours-per-week','native-country','class']
data.columns = columns
```

```python
data.info()
```

```python
data['class'].value_counts()
```

# Data Cleaning

```python
df = data[(data['workclass']!=' ?') & (data['education']!=' ?') & (data['marital-status']!=' ?') & (data['occupation']!=' ?') & 
        (data['relationship']!=' ?') & (data['race']!=' ?') & (data['sex']!=' ?') & (data['native-country']!=' ?')]
```

```python
df.head()
```

# Visualizations

```python
plt.rcParams["figure.figsize"]=(5, 4)
sns.histplot(data=df,x='age',hue='class',bins=20,kde=True)
plt.show()
sns.boxplot(data=df,x='age',y='class',orient='h')
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(13, 10)
mosaic(df,['workclass','class'])
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(5, 4)
sns.histplot(data=df,x='fnlwgt',hue='class',bins=20,kde=True)
plt.show()
sns.boxplot(data=df,x='fnlwgt',y='class',orient='h')
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(15, 10)
mosaic(df,['education','class'])
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(8, 5)
sns.countplot(data=df,x='marital-status',hue='class')
plt.xticks(rotation=90)
plt.show()
```
